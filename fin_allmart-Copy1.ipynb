```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from time import sleep
import pandas as pd
import re
options = webdriver.ChromeOptions()
options.add_argument("--incognito")
```

---------

```python
# pxmart 全聯
# get both cities at the same time
driver = webdriver.Chrome(ChromeDriverManager().install())
pxList = []

def visit():
    driver.get("http://www.pxmart.com.tw/px/map.px")
    sleep(8)
    
def get_addr():
    menu = driver.find_element_by_css_selector("div > #ipt-city")
    hidden =  driver.find_element_by_css_selector("div > #ipt-city > a")    
    
    for i in range(4,6): # tpe & ntp city
        link='#ipt-city > ul > li:nth-child('+str(i)+') > button'
        ActionChains(driver).move_to_element(menu).click(hidden).perform()
        driver.find_element(By.CSS_SELECTOR, link).click() 
        sleep(3)
        driver.find_element(By.CSS_SELECTOR,'a#btn-search').click() # click search
        sleep(3)
        addr=driver.find_elements(By.CSS_SELECTOR,'#paxmart-map-results > li> div.paxmart-map-results-info > p:nth-child(5)') #各店地址
        for a in addr:
            address=a.text
            pxList.append(address)
        driver.find_element(By.CSS_SELECTOR,'div.paxmart-map-option-oc-btn > div > a').click()
        sleep(3)

    print(pxList)

visit()
get_addr()
```

```python
# simplemart 美廉社
# get both cities at the same time
driver = webdriver.Chrome(ChromeDriverManager().install())
sList = []

def visit():
    driver.get("https://www.simplemart.com.tw/ec99/ushop20097/map.asp")
    sleep(8)
def get_addr():   
    driver.find_element(By.CSS_SELECTOR,'div>div>#btn-search').click() # click search 按下搜尋
    sleep(10)

    addr=driver.find_elements(By.CSS_SELECTOR,'div>div >#plF > ul > li.map-add') # get address 取得地址
    for a in addr:
        address=a.text
        if '台北市' in address:
            sList.append(address)
        elif '新北市' in address:
            sList.append(address)
    print(sList)
visit()
get_addr()
```

```python
# wellcome 頂好: ntp 新北
driver = webdriver.Chrome(ChromeDriverManager().install())
wList = []
j=0
def visit():
    driver.get("http://www.wellcome.com.tw/CHT/HOME/Store")
    sleep(8)
    
def get_ntp_addr():   
    driver.find_element(By.CSS_SELECTOR,'#City > option:nth-child(3)').click() # click ntp
    sleep(3)
    driver.find_element(By.CSS_SELECTOR,'#Search > li:nth-child(1) > input[type=button]').click() # click search
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located( 
                (By.CSS_SELECTOR, "address")   
            )
        )
        print('found')
    except TimeoutException:
        print("error")
        
def parse():
    global j
    sleep(8)
    addr=driver.find_elements(By.CSS_SELECTOR,'div>#SearchResult > div.result > dl > address') #各店地址
    
    for a in addr:
        address=a.text
        print(address)
        wList.append(address)
    
    if j < 5: # go thru last 5 pages
        if len(driver.find_elements(By.CSS_SELECTOR, '#divPaging > ul > li:nth-child(8) > a') ) > 0:
            driver.find_element(By.CSS_SELECTOR, '#divPaging > ul > li:nth-child(8) > a').click()
#             print(j)
            j+=1
            parse()
visit()
get_ntp_addr()
parse()
```

```python
# wellcome 頂好: tpe 台北
driver = webdriver.Chrome(ChromeDriverManager().install())
# wList= []
j=0
def visit():
    driver.get("http://www.wellcome.com.tw/CHT/HOME/Store")
    sleep(8)
    
def get_tpe_addr():   
    driver.find_element(By.CSS_SELECTOR,'#City > option:nth-child(2)').click() # click tpe
    sleep(3)
    driver.find_element(By.CSS_SELECTOR,'#Search > li:nth-child(1) > input[type=button]').click() # click search
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located( 
                (By.CSS_SELECTOR, "address")   
            )
        )
        print('found')
    except TimeoutException:
        print("error")
        
def parse():
    global j
    sleep(8)
    addr=driver.find_elements(By.CSS_SELECTOR,'div>#SearchResult > div.result > dl > address') #各店地址
    
    for a in addr:
        address=a.text
#         print(address)
        wList.append(address)
    
    if j < 4: # go thru last 4 pages
        if len(driver.find_elements(By.CSS_SELECTOR, '#divPaging > ul > li:nth-child(7) > a') ) > 0:
            driver.find_element(By.CSS_SELECTOR, '#divPaging > ul > li:nth-child(7) > a').click()
#             print(j)
            j+=1
            parse()


visit()
get_tpe_addr()
parse()
print(wList)
```
